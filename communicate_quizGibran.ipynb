```python
# imports and load data
import pandas as pd


df= pd.read_csv('store_data.csv')

df.head()
```

```python
# explore data
df.describe()
df.info()
```

```python
# sales for the last month
df['week']=pd.to_datetime(df['week'])
df.set_index(df['week'],inplace=True)
g = df.groupby(pd.TimeGrouper("M"))
sumg=g.sum()
sumg.tail(1).plot(kind='bar');
```

```python
# average sales
df.mean().plot(kind='bar');
```

```python
# sales for the week of March 13th, 2016
march13=df[df.week =='2016-03-13']
march13.plot(kind='bar')
```

```python
# sales for the lastest 3-month periods
sumg=g.sum()
sumg.tail(3).sum().plot(kind='bar');
```
